**STRING** --- sequence of one or more characters , writen in btw ("") or ('')..

1) Indexing---starts from 0

```python
text = "PYTHON"
print(text[0])
print(text[2])
print(text[5])
print(text[-3])
```

2.String Slicing --- stringname[start:end]

```python
text = "Python"

print(text[0:3])
print(text[2:5])
print(text[:4])
print(text[3:])
```

3.String legth --- len(stringname)

```python
text = "Python"

print(len(text))
```

5. String Methods

5.1) Upper case-- strname.upper()

```python
text = "python"
print(text.upper())
```

5.2) Lower case-- strname.lower()

```python
text = "PYTHON"
print(text.lower())
```

5.3)Capitalize -- strname.capitalize()

```python
text = "python programming"
print(text.capitalize())
```

5.4)Replace-- strname.replace("//original","//replaced")

```python
text = "I like Java"
print(text.replace("Java","Python"))
```

5.5)Find--strname.find(//char which needs to find)

```python
text = "Python programming"
print(text.find("pro"))
```

6.String Concatenation--to combine two strings

```python
first = "Hello"
second = "World"

print(first + " " + second)
```

7.String Repetition

```python
print("pythob " * 3)
```

8. Check in String -- output boolion

```python
text = "Python programming"

print("Python" in text)
```

**LISTS**-- a collection of multiple items stored in a single variable. written btw [ ]

```python
numbers = [10, 20, 30, 40]
names = ["Ram", "Sam", "John"]
mixed = [10, "Python", 3.5]
print(numbers)
print(names)
print(mixed)
```

1.Indexing

```python
fruits = ["apple", "banana", "mango"]
print(fruits[0])
print(fruits[2])
print(fruits[-1])
```

2. List Slicing -- used to get a part of the list.--list[start:end]

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
print(numbers[3:])
print(numbers[:4])
```

3. Changing List Elements-- List is mutable

```python
fruits = ["apple", "banana", "mango"]

fruits[1] = "orange"

print(fruits)
```

4. Adding Elements to a List - there are two ways

4.1) append ()--add an element at the end of the list

```python
numbers = [10, 20, 30]
numbers.append(40)
print(numbers)
```

4.2) insert() -- Adds an element at a specific position.--listname.insert(position, element)

```python
numbers = [10, 20, 30]

numbers.insert(1, 15)

print(numbers)
```

5) Removing Elements -- there are two ways

5.1) remove()--Removes a specific value.

```python
numbers = [10, 20, 30, 40]

numbers.remove(30)

print(numbers)
```

5.2)pop()---Removes element using index position

```python
numbers = [10, 20, 30]

numbers.pop(1)

print(numbers)
```

6) List Length -- Use len() to find the number of elements

```python
numbers = [10, 20, 30, 40]

print(len(numbers))
```

7) Looping Through a List --- Using for loop:

```python
numbers = [10, 20, 30]

for i in numbers:
    print(i)
```


8) List Concatenation -- Lists can also be joined using +.

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = list1 + list2

print(result)
```

**ARRAYS** ---it is a data structure used to store multiple elements of the same data type in a single variable.

1) Creating an array -- First, import the array module.-- import array as arr

```python
import array as arr

numbers = arr.array('i', [10, 20, 30, 40])
print(numbers)
```

2. Accessing Array Elements-- INDEXING


```python
import array as arr

numbers = arr.array('i', [10, 20, 30, 40])

print(numbers[0])
print(numbers[2])
```

3. Adding Elemens to Array -- there are two ways

3.1) append()-- Adds an element to the end of the array.

```python
import array as arr

numbers = arr.array('i', [10, 20, 30, 40])
numbers.append(50)
print(numbers)
```

3.2)insert()--Adds an element at a specific index.


```python
import array as arr

numbers = arr.array('i', [10, 20, 30, 40])
numbers.insert(1, 15)
print(numbers)
```

4. Removing Elements -- there are two ways

4.1) remove()--removes a specific values

```python
import array as arr

numbers = arr.array('i', [10, 20, 30, 40])
numbers.remove(30)
print(numbers)
```

4.2)pop()--Removes element using index.

```python
import array as arr

numbers = arr.array('i', [10, 20, 30, 40])
numbers.pop(1)
print(numbers)
```

5. Traversing an Array (Loop)--You can print all elements using a for loop.

```python
import array as arr

numbers = arr.array('i', [10, 20, 30, 40])

for i in numbers:
    print(i)
```
