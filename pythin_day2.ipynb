**Tuples**--a collection of multiple items stored in a single variables , similar to list , which are immutable


1. Creating a Tuple

```python
numbers = (10, 20, 30, 40)
names = ("Ram", "Sam", "John")
mixed = (10, "Python", 3.5)
print (numbers)
print (names)
print (mixed)
```

2)Indexing a tuple


```python
fruits = ("apple", "banana", "mango")

print(fruits[0])
print(fruits[2])
print(fruits[-1])
```

3)Tuple Slicing--You can get a portion of a tuple using slicing.-- tuple[start:end]

```python
numbers = (10, 20, 30, 40, 50)

print(numbers[1:4])
```

4. Tuple Length--Use len() to find the number of elements.

```python
numbers = (10, 20, 30, 40)

print(len(numbers))
```

5. Looping Through a Tuple--You can use a for loop.

```python
numbers = (10, 20, 30)

for i in numbers:
    print(i)
```

6. Tuple Concatenation---Tuples can be joined using the + operator.

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

result = tuple1 + tuple2

print(result)
```

7. Tuple Repetition---You can repeat tuples using *.

```python
numbers = (1, 2)
print(numbers * 3)
```
